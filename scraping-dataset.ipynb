```python
!pip install google-play-scraper
from google_play_scraper import reviews, Sort
import pandas as pd

apps = ['com.shopee.id']
all_reviews = []

for app_id in apps:
    scraped, _ = reviews(
        app_id,
        lang='id',
        country='id',
        sort=Sort.MOST_RELEVANT,
        count=15000  
    )
    all_reviews.extend(scraped)

df_busines = pd.DataFrame(all_reviews)

df_busines.drop_duplicates(subset=['content'], inplace=True)
df_busines = df_busines[df_busines['content'].notna()]

print(f"Total data terkumpul: {len(df_busines)} sampel")

df_busines.to_csv('dataset.csv', index=False)

df_busines.head(5)
```
